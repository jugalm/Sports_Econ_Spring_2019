## Read a csv file using pandas package

```python
import pandas as pd

df = pd.read_csv('/Users/jugalmarfatia/docs/college_draft.csv')
```

```python
df.head()
```

## Top schools with the highest number of nfl drafts since 2000

In order to find that we need find total count by each school

```python
df[['SCHOOL', 'PLAYER']].groupby(['SCHOOL']).count().sort_values('PLAYER', ascending = False)
```

## Code to list all NFL drafts from wsu

```python
df[df['SCHOOL'] == "Washington State"]
```

## We can also create a new dataframe with the above table

```python
df_washington = df[df['SCHOOL'] == "Washington State"]
df_washington
```

## We can do the same for ohio
## Additional we can count the number of NFL drafts each year

```python
df_ohio = df[df['SCHOOL'] == "Ohio State"]

df_ohio_by_year = df_ohio[['year', 'PLAYER']].groupby(['year']).count().reset_index()

df_ohio_by_year
```

```python
import matplotlib.pyplot as plt

plt.plot(df_ohio_by_year['year'], df_ohio_by_year['PLAYER'], '.-')
```
